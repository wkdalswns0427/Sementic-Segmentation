```python
import os
import PIL.Image as img
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readBMP(img_dir):
    outimg = []
    images = os.listdir(img_dir)
    for image in images:
        img_path = os.path.join(img_dir, image)
        rawimg = img.open(img_path)
        rawimg = np.array(rawimg)
        outimg.append(rawimg)
    return outimg
```

```python
IMG_DIR = "ChallengeSample/"
```

```python
Image = readBMP(IMG_DIR)
```

```python
plt.imshow(Image[0], cmap='gray')
```

```python
for i in range(len(Image)):
    Image[i][Image[i]<55]=0
    Image[i][Image[i]>54]=255

plt.imshow(Image[0], cmap='gray')
```

```python
[R,C] = Image[1].shape
print(R, C)
```

```python
B, W = 0, 0
Pos = []
for j in range(C):
    for k in range(R):
        if Image[1][k][j]==0 :
            W = W+1
        else:
            B = B+1
    Ratio = W/(B+W)
    if Ratio>0.8:
        Pos.append(j)
```

```python
print(Pos)
```

```python
axis1=[]
axis2=[]
for p in Pos:
    for k in range(0,R-3):
        if (Image[1][k-3][p]==0 and Image[1][k][p]==0 and Image[1][k+3][p]==0):
            axis1.append(k)
            axis2.append(p)
```

```python
plt.ylim([0,R])
plt.xlim([0,len(Pos)])
plt.scatter(axis2,axis1, c='black', s=0.1)
plt.show()

```
